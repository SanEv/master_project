```python
import numpy as np
import pandas as pd
import os
import glob
```

```python
folderpath = os.path.dirname(r'C:\Users\visan\Downloads\days_test\\')
zenith_file_name = '??????_K.NO2_vis'
horizon_file_name = '??????KK.NO2_vis_sync'
zenith_filepaths = glob.glob(os.path.join(folderpath,zenith_file_name))
horizon_filepaths = glob.glob(os.path.join(folderpath,horizon_file_name))

print(zenith_filepaths)
print(horizon_filepaths)

 
```

```python
def read_file_columns(fle, skip_line, cols):
    if os.stat(fle).st_size == 0:
        raise ValueError("File is empty")
    with open(fle) as f:
        pos = 0
        cur_line = f.readline()
        while not cur_line.startswith(skip_line):
            pos = f.tell()
            cur_line = f.readline()
        f.seek(pos)
        return pd.read_csv(f,sep='\s+',error_bad_lines=False, usecols = cols)
```

```python
for i in zenith_filepaths:
    zenith_data = read_file_columns(i, "*day-of-year", [1,6,10,12,16,30])
    print(zenith_data)
    
for i in horizon_filepaths:
    horizon_data = read_file_columns(i, "*day-of-year", [1,6,13,15,19,33])
    print(horizon_data)
```
